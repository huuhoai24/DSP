```python


from datasets import load_dataset

# Tải dataset
dataset = load_dataset("PB3002/ViMedical_Disease")

# Xem thử 1 dòng dữ liệu
print(dataset['train'][0])

# Chuyển sang dạng Pandas DataFrame để dễ xử lý
df = dataset['train'].to_pandas()
print(df.head())
```

```python
import pandas as pd
disease = pd.read_csv("disease_department_mapping.csv")
disease
```

```python
df["medical_specialty"] = df["Disease"].map(disease.set_index("disease_name")["department"])
```

```python
df = df[df["medical_specialty"].notna()]
```

```python
df.to_csv("ViMedical_Disease.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 10)

print("=" * 80)
print("EXPLORATORY DATA ANALYSIS - ViMedical Dataset")
print("=" * 80)

# 1. Dataset Shape & Basic Info
print("\n1. DATASET SHAPE & BASIC INFO")
print(f"   Total rows: {df.shape[0]}")
print(f"   Total columns: {df.shape[1]}")
print(f"   Columns: {list(df.columns)}")

# 2. Data Types
print("\n2. DATA TYPES")
print(df.dtypes)

# 3. Missing Values
print("\n3. MISSING VALUES")
missing = df.isnull().sum()
print(missing[missing > 0] if missing.sum() > 0 else "   No missing values!")

# 4. Disease Distribution
print("\n4. DISEASE DISTRIBUTION")
disease_counts = df["Disease"].value_counts()
print(f"   Unique diseases: {df['Disease'].nunique()}")
print(f"\n   Top 10 diseases:")
print(disease_counts.head(10))

# 5. Medical Specialty Distribution
print("\n5. MEDICAL SPECIALTY DISTRIBUTION")
specialty_counts = df["medical_specialty"].value_counts()
print(f"   Unique specialties: {df['medical_specialty'].nunique()}")
print(f"\n   All specialties:")
print(specialty_counts)

# 6. Text Analysis
df["Question_Length"] = df["Question"].str.len()
df["Question_WordCount"] = df["Question"].str.split().str.len()

print("\n6. QUESTION TEXT ANALYSIS")
print(f"   Avg question length: {df['Question_Length'].mean():.2f} characters")
print(f"   Max question length: {df['Question_Length'].max()} characters")
print(f"   Min question length: {df['Question_Length'].min()} characters")
print(f"   Avg words per question: {df['Question_WordCount'].mean():.2f}")

# 7. Visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Top diseases
top_diseases = df["Disease"].value_counts().head(10)
axes[0, 0].barh(range(len(top_diseases)), top_diseases.values)
axes[0, 0].set_yticks(range(len(top_diseases)))
axes[0, 0].set_yticklabels(top_diseases.index)
axes[0, 0].set_xlabel("Count")
axes[0, 0].set_title("Top 10 Diseases")
axes[0, 0].invert_yaxis()

# Plot 2: Medical specialists distribution
specialty_counts.plot(kind='barh', ax=axes[0, 1])
axes[0, 1].set_xlabel("Count")
axes[0, 1].set_title("Medical Specialty Distribution")

# Plot 3: Question length distribution
axes[1, 0].hist(df["Question_Length"], bins=30, color='skyblue', edgecolor='black')
axes[1, 0].set_xlabel("Question Length (characters)")
axes[1, 0].set_ylabel("Frequency")
axes[1, 0].set_title("Distribution of Question Lengths")

# Plot 4: Question word count distribution
axes[1, 1].hist(df["Question_WordCount"], bins=30, color='lightcoral', edgecolor='black')
axes[1, 1].set_xlabel("Word Count")
axes[1, 1].set_ylabel("Frequency")
axes[1, 1].set_title("Distribution of Question Word Count")

plt.tight_layout()
plt.show()

# 8. Statistics Summary
print("\n7. STATISTICS SUMMARY")
print(df[["Question_Length", "Question_WordCount"]].describe())
```
